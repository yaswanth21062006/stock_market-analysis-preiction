```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
stocks =["HINDCOPPER.NS","KOTAKBANK.NS","SBIN.NS","TATACONSUM.NS","ASIANPAINT.NS"]
data =  yf.download(stocks,start="2015-01-01",end="2025-01-01")
data.head(10)
```

```python
closing_prices = data.Close
closing_prices.head()
```

```python
yearly_prices = closing_prices.resample('Y').first()

yearly_prices.head(11)
```

```python
yearly_prices.index =  yearly_prices.index.strftime('%Y')
plt.figure(figsize=(40,20))

yearly_prices.plot(kind='bar')
plt.title("Stock Price Analysis")
years =[2015,2016,2017,2018,2019,2020,2021,2022,2023,2024]

plt.xlabel("Years")
plt.ylabel("Prices")


plt.show()
```

Stock Price Comparison Analysis

The bar graph shows the yearly stock price comparison of the selected companies over the 10-year period.

From the graph, it can be observed that Asian Paints consistently has the highest stock price among all the companies. The price shows strong growth over the years, indicating stable market performance.

Hindustan Copper has the lowest stock price among the selected companies. However, its price shows noticeable fluctuations over the years, indicating higher volatility.

The banking stocks Kotak Mahindra Bank and State Bank of India show moderate growth trends during the period. Their prices increase gradually compared to earlier years.

Tata Consumer Products also shows a steady upward trend, indicating consistent growth in the consumer goods sector.

Overall, the comparison shows that companies from different sectors exhibit different growth patterns in the stock market.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
closing_prices=closing_prices.reset_index(drop=True)
closing_prices.head()
closing_prices['Prev_Close'] = closing_prices['ASIANPAINT.NS'].shift(1)
closing_prices = closing_prices.dropna()
y = closing_prices['ASIANPAINT.NS']
x = closing_prices[['Prev_Close']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2,shuffle=False) 
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
predicted = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(Y_test, predicted)
print("Model Accuracy :",accuracy)
```

```python
plt.figure(figsize=(10,5))
plt.plot(Y_test.values, label='Actual Price')
plt.plot(predicted, label="Predicted Price")
plt.title("Asain PaintsStock price pediction using Linear Regression")
plt.xlabel("Days")
plt.ylabel("Stock Price")

plt.legend()
plt.show()
         
```

```python
results = pd.DataFrame({"Actual Price" : Y_test.values,
                        "Predicted Price" : predicted})
results.head()
```
